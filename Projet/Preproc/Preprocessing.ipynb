# Target Analysis

Le lien du dataset nous montre que l'on peut consifÃ©der CL0 ("never used") et CL1 ("used a decade ago") as non user people and the other as user people

```python
df.replace("CL0", 0, inplace = True)
df.replace("CL1", 0, inplace = True)
df.replace("CL2", 1, inplace = True)
df.replace("CL3", 1, inplace = True)
df.replace("CL4", 1, inplace = True)
df.replace("CL5", 1, inplace = True)
df.replace("CL6", 1, inplace = True)
```

# Clustering levels of dependency

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(df)
    print (i,kmeans.inertia_)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('WSS')
plt.savefig('../../image/Elbow_knn.png',bbox_inches='tight')
plt.show()
```

Nombre de cluster ideal = 3

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',random_state =10)
estimator = kmeans.fit(df)
estimator
```

```python
class_predict = estimator.predict(df)
class_predict
```

```python
df["Addiction"] = class_predict
df
```
