# Multivariate analysis

```python
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams.update({'font.size': 15})
fig, ax = plt.subplots(1, 1, figsize=(20,15))
sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, ax=ax, cmap=plt.cm.Greens)
plt.show()
```

## Scores

### Nscore

```python
score = pd.DataFrame()
score["Amphet"] = df[df["Amphet"]==1]["Nscore"]
score["Amyl"] = df[df["Amyl"]==1]["Nscore"]
score["Benzos"] = df[df["Benzos"]==1]["Nscore"]
score["Cannabis"] = df[df["Cannabis"]==1]["Nscore"]
score["Coke"] = df[df["Coke"]==1]["Nscore"]
score["Ecstasy"] = df[df["Ecstasy"]==1]["Nscore"]
score["Heroin"] = df[df["Heroin"]==1]["Nscore"]
score["Ketamine"] = df[df["Ketamine"]==1]["Nscore"]
score["Legalh"] = df[df["Legalh"]==1]["Nscore"]
score["LSD"] = df[df["LSD"]==1]["Nscore"]
score["Meth"]  = df[df["Meth"]==1]["Nscore"]
score["Mushrooms"] = df[df["Mushrooms"]==1]["Nscore"]
score["Nicotine"] = df[df["Nicotine"]==1]["Nscore"]
score["VSA"] = df[df["VSA"]==1]["Nscore"]
score["Crack"] = df[df["Crack"]==1]["Nscore"]
```

```python
score
```

```python
# Boxplot NScore vs who takes Drugs

plt.rcParams['axes.labelsize'] = 20
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 18
plt.rcParams['font.size'] = 26
plt.figure(figsize=(10, 7))

plt.title('Nscore and Drugs', fontsize=20)

ax = sns.boxplot(data=score,orient="h", palette="Set2")

plt.ylabel('Nscore')
```

Looking at the fist analysis of Nscore and the target we get that level of Addiction 2 was negative, quite similar but more around score 0 we get the level 0 and level 1 was bigger than both and positive.
