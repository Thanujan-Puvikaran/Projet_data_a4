# Model

```python
# Split train test

X_train, X_test, y_train, y_test = train_test_split(df.drop("Addiction", axis = 1), df["Addiction"], test_size=0.35, random_state = 42,\
                                                   stratify = df["Addiction"])
```

```python
# train dataset

train_dataset = pd.concat([X_train, y_train],axis=1)
train_dataset.head()
```

```python
# hyperparameters

parameters = {'bootstrap': [True, False],\
 'max_depth': [2, 3, 4, 5, 6, 7, 8],\
 'max_features': ['auto', 'sqrt'],\
 'min_samples_leaf': [2, 3, 4, 5, 6, 7, 8, 9],\
 'min_samples_split': [2, 3, 4, 5, 6, 7, 8, 9],\
 'n_estimators': [50, 100, 150, 200, 250, 300]}
```

```python
# Using kfold strat

kfold = StratifiedKFold(n_splits=3, random_state=42, shuffle = True)
```

```python
#RandomizedSearchCV

class_random = RandomizedSearchCV(RandomForestClassifier(),\
parameters, cv=kfold, scoring=['recall', 'precision', 'f1', 'roc_auc', 'accuracy'], refit='accuracy', n_iter=300)
```
