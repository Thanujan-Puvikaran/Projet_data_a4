# Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import minmax_scale, RobustScaler,QuantileTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, plot_confusion_matrix, f1_score
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
from sklearn.model_selection import train_test_split
from IPython.display import HTML, display
import warnings
from sklearn.ensemble import RandomForestClassifier
```

# Dataset

```python
df = pd.read_csv("../Models/data_tofit.csv")
df
```

# Model

```python
# Split train test

X_train, X_test, y_train, y_test = train_test_split(df.drop("Addiction", axis = 1), df["Addiction"], test_size=0.35, random_state = 42,\
                                                   stratify = df["Addiction"])
```

```python
# train dataset

train_dataset = pd.concat([X_train, y_train],axis=1)
train_dataset.head()
```

```python
train_dataset.shape
```

```python
# Hyperparameters
param = {"max_depth": [2,6,8,10] ,\
        "min_samples_leaf": [1,2,5,10,15],\
        "n_estimators": [500,800,1500,2500,5000],\
        "max_features":["auto", "sqrt","log2"]}
```

```python
grid_search = GridSearchCV(RandomForestClassifier(), param, n_jobs=-1, cv=5, scoring = ["accuracy"],refit=False)
grid_search.fit(X_train, y_train)
```

```python
pd.DataFrame(grid_search.cv_results_).sort_values("mean_test_accuracy", ascending = False)
```

```python
grid_search_withoutscoring = GridSearchCV(RandomForestClassifier(), param, n_jobs=-1, cv=5)
grid_search_withoutscoring.fit(X_train, y_train)
```

```python
print("Best parameters {} with best score {} ".format(grid_search_withoutscoring.best_params_,\
grid_search_withoutscoring.best_score_))
```

```python
# Hyperparameters
param2 = {"max_depth": [10,20,30,40,50] ,\
        "min_samples_leaf": [1,2,5,10,15],\
        "n_estimators": [500,800,1500,2500,5000],\
        "max_features":["auto", "sqrt","log2"]}
```

```python
grid_search_withoutscoring2 = GridSearchCV(RandomForestClassifier(), param2, n_jobs=-1, cv=5)
grid_search_withoutscoring2.fit(X_train, y_train)
```

```python
print("Best parameters {} with best score {} ".format(grid_search_withoutscoring2.best_params_,\
grid_search_withoutscoring2.best_score_))
```

## GridSearchCV without scoring + StratifiedKFold

```python
kfold = StratifiedKFold(n_splits=5, random_state=42, shuffle = True)
```

```python
grid_search_withoutscoring = GridSearchCV(RandomForestClassifier(), param, n_jobs=-1, cv=kfold)
grid_search_withoutscoring.fit(X_train, y_train)
```

```python
print("Best parameters {} with best score {} ".format(grid_search_withoutscoring.best_params_,\
grid_search_withoutscoring.best_score_))
```

```python
# Importance Featuring

importance = pd.DataFrame(data=grid_search_withoutscoring.best_estimator_.feature_importances_, columns = ['Value'],\
index=X_test.columns).reset_index().rename(columns={'index':'Variable'}).sort_values('Value').reset_index(drop=True)
importance['%'] = importance['Value']*100
importance
```

```python
# Model importance
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 15
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(10,7))
plt.barh(importance['Variable'], importance['Value'])
```

## Remove more features

```python
df_important2 = df_important.drop(columns = ["Education", "Age", "Ascore", "LSD"], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_important2.drop("Addiction", axis = 1), df_important2["Addiction"], test_size=0.35, random_state = 42,\
                                                   stratify = df_important2["Addiction"])
```

```python
kfold = StratifiedKFold(n_splits=5, random_state=42, shuffle = True)
```

```python
grid_search_important2= GridSearchCV(RandomForestClassifier(), param, n_jobs=-1, cv=kfold)
grid_search_important2.fit(X_train, y_train)
```

```python
print("Best parameters {} with best score {} ".format(grid_search_important2.best_params_,\
grid_search_important2.best_score_))
```

```python
# Importance Featuring

importance = pd.DataFrame(data=grid_search_important2.best_estimator_.feature_importances_, columns = ['Value'],\
index=X_test.columns).reset_index().rename(columns={'index':'Variable'}).sort_values('Value').reset_index(drop=True)
importance['%'] = importance['Value']*100
importance
```

```python
# Model importance
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 15
plt.rcParams.update({'font.size': 20})   
plt.figure(figsize=(10,7))
plt.barh(importance['Variable'], importance['Value'])
```

```python
# Testing model


y_pred = grid_search_important2.predict(X_test)
f1_score(y_test, y_pred)
```

```python
# Montando a matriz de confusão

cross = pd.crosstab(y_test, y_pred, rownames=['Real'], colnames=['Predito'], margins=True)
cross
```

```python
# Olhando a matriz de confusão percentual

plt.rcParams['axes.labelsize'] = 20
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams.update({'font.size': 22})
fig, ax = plt.subplots(1, 1, figsize=(10,7))
fig.suptitle('Matriz de Confusão Normalizada', fontsize=15)
plot_confusion_matrix(grid_search_important2, X_test, y_test, normalize='true', cmap=plt.cm.Blues, ax=ax)
plt.xlabel('Classe Predita')
plt.ylabel('Classe Real')
plt.show() 
```
