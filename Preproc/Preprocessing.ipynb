```python
# j'enl√®ve les colonnes avec des nan
df.dropna(inplace = True) 
df.isnull().any()
```

```python
colonne = list()
with open('map_coluns.txt') as f:
    lines = f.readlines()
    colonne = lines
```

```python
colonne = [re.sub("X\d{1,3} ","",x) for x in colonne]
colonne = [re.sub("\n","",x) for x in colonne]
colonne.append("class")
```

```python
df.set_axis(colonne, axis=1, inplace=True)
df
```

# Target treatment

```python
#on voit les diff√©rentes valeurs de la target
df["class"].unique()
```

```python
#nb valeur de la classe 0
nbc1 = len(df[df["class"] == df["class"].unique()[0]])
nbc1
```

```python
#nb valeur de la classe 1
nbc2 = len(df[df["class"] == df["class"].unique()[1]])
nbc2
```

```python
Pourcentage1 = len(df[df["class"] == df["class"].unique()[0]])/len(df)
Pourcentage1
```

```python
Pourcentage2 = len(df[df["class"] == df["class"].unique()[1]])/len(df)
Pourcentage2
```

```python
#Mauvais equilibre, attention au mod√®le choisi
```

```python
#On d√©finit les classes de la target
df.loc[df["class"] == df["class"].unique()[0], "bankrupt"] = 0
df.loc[df["class"] == df["class"].unique()[1], "bankrupt"] = 1
df["bankrupt"] =df["bankrupt"].astype(int) 
del df["class"]
df.head()
```
